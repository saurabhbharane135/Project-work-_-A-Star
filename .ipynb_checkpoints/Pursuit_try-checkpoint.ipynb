```python
#Pure Pursuit Final, Reaches the goal and stops and python closes the script
import sim
import math
import time
import sys  # Import the sys module

# Connect to CoppeliaSim
sim.simxFinish(-1)
clientID = sim.simxStart('127.0.0.1', 19999, True, True, 5000, 5)

if clientID != -1:
    print('Connected Successfully.')
else:
    print('Connection Failed.')
    exit()

# Get robot handles
error_code, left_motor_handle = sim.simxGetObjectHandle(clientID, '/PioneerP3DX/leftMotor', sim.simx_opmode_oneshot_wait)
error_code, right_motor_handle = sim.simxGetObjectHandle(clientID, '/PioneerP3DX/rightMotor', sim.simx_opmode_oneshot_wait)

# Define path coordinates
path = [
    [0.5, 0],
    [1.0, 0],
    [1.5, 0],
    [1.5, 0.5],
    [1.5, 0.5]
    
]

# Pure pursuit control parameters
lookahead_distance = 0.5  # Adjust as needed
linear_velocity = 0.5    # Adjust as needed
goal_radius = 0.1        # Adjust as needed

# Function to get robot pose
def get_robot_pose():
    error_code, robot_position = sim.simxGetObjectPosition(clientID, left_motor_handle, -1, sim.simx_opmode_streaming)
    error_code, robot_orientation = sim.simxGetObjectOrientation(clientID, left_motor_handle, -1, sim.simx_opmode_streaming)
    return robot_position, robot_orientation[2]  # Extract yaw angle

# Pure pursuit implementation
def pure_pursuit():
    # Add a flag to track if the final goal has been reached
    if hasattr(pure_pursuit, 'final_goal_reached') and pure_pursuit.final_goal_reached:
        sys.exit()  # Exit the script when final goal is reached

    robot_position, robot_yaw = get_robot_pose()

    # Find nearest waypoint on path
    nearest_index = 0
    nearest_distance = float('inf')
    for i, waypoint in enumerate(path):
        distance = math.hypot(waypoint[0] - robot_position[0], waypoint[1] - robot_position[1])
        if distance < nearest_distance:
            nearest_index = i
            nearest_distance = distance

    # Check if reached goal
    if nearest_distance <= goal_radius:
        # Update goal to the next waypoint
        goal_index = min(nearest_index + 1, len(path) - 1)
        goal_point = path[goal_index]

        # Check if the last waypoint is reached
        if goal_index == len(path) - 1:
            # Stop motors only if the final goal hasn't been reached yet
            if not hasattr(pure_pursuit, 'final_goal_reached') or not pure_pursuit.final_goal_reached:
                sim.simxSetJointTargetVelocity(clientID, left_motor_handle, 0, sim.simx_opmode_oneshot)
                sim.simxSetJointTargetVelocity(clientID, right_motor_handle, 0, sim.simx_opmode_oneshot)
                print("Reached the final goal!")
                pure_pursuit.final_goal_reached = True  # Set the flag to indicate final goal reached
                sys.exit()  # Exit the script when final goal is reached
            return  # Exit the pure_pursuit function

    else:
        goal_point = path[nearest_index]

    # Calculate look-ahead point
    look_ahead_point = path[min(nearest_index + 1, len(path) - 1)]

    # Calculate steering angle
    alpha = math.atan2(look_ahead_point[1] - robot_position[1], look_ahead_point[0] - robot_position[0]) - robot_yaw

    # Set motor velocities
    angular_velocity = alpha * linear_velocity / lookahead_distance
    left_velocity = linear_velocity - angular_velocity * 0.5 * 0.3  # Adjust wheel separation if needed
    right_velocity = linear_velocity + angular_velocity * 0.5 * 0.3
    sim.simxSetJointTargetVelocity(clientID, left_motor_handle, left_velocity, sim.simx_opmode_oneshot)
    sim.simxSetJointTargetVelocity(clientID, right_motor_handle, right_velocity, sim.simx_opmode_oneshot)

# Main loop
while True:
    pure_pursuit()
    time.sleep(0.05)  # Adjust the time interval as needed
```

```python
import sim
import math
import time
import sys

# Connect to CoppeliaSim
sim.simxFinish(-1)
clientID = sim.simxStart('127.0.0.1', 19999, True, True, 5000, 5)

if clientID != -1:
    print('Connected Successfully.')
else:
    print('Connection Failed.')
    exit()

# Set a fixed time step for simulation
sim.simxSetFloatingParameter(clientID, sim.sim_floatparam_simulation_time_step, 0.05, sim.simx_opmode_oneshot)

# Get robot handles
error_code, left_motor_handle = sim.simxGetObjectHandle(clientID, '/PioneerP3DX/leftMotor', sim.simx_opmode_oneshot_wait)
error_code, right_motor_handle = sim.simxGetObjectHandle(clientID, '/PioneerP3DX/rightMotor', sim.simx_opmode_oneshot_wait)

# Define path coordinates
path = [
    [-1.75, 2.25],
    [-1.6, 2.25],
    [-1.45, 2.25],
    [-1.4, 2.25],
    [-1.35, 2.25],
    [-1.25, 2.25],
    [-1.25, 1.75],
    [-1.25, 1.25],
    [-1.25, 0.75],
    [-1.25, 0.25],
    [-1.25, -0.25],
    [-1.25, -0.75],
    [-1.25, -1.25],
    [-1.25, -1.75],
    [-0.75, -1.75],
    [-0.25, -1.75],
    [-0.25, -1.25],
    [-0.25, -0.75],
    [-0.25, -0.25],
    [0.25, -0.25],
    [0.75, -0.25],
    [1.25, -0.25],
    [1.75, -0.25],
    [1.75, 0.25]
]

# Pure pursuit control parameters
lookahead_distance = 1.0  # Adjust as needed
linear_velocity = 2.0    # Adjust as needed
goal_radius = 0.2        # Adjust as needed

# Function to get robot pose
def get_robot_pose():
    error_code, robot_position = sim.simxGetObjectPosition(clientID, left_motor_handle, -1, sim.simx_opmode_streaming)
    error_code, robot_orientation = sim.simxGetObjectOrientation(clientID, left_motor_handle, -1, sim.simx_opmode_streaming)
    return robot_position, robot_orientation[2]  # Extract yaw angle

# Pure pursuit implementation
def pure_pursuit():
    # Add a flag to track if the final goal has been reached
    if hasattr(pure_pursuit, 'final_goal_reached') and pure_pursuit.final_goal_reached:
        sys.exit()  # Exit the script when the final goal is reached

    robot_position, robot_yaw = get_robot_pose()

    # Find the nearest waypoint on the path
    nearest_index = 0
    nearest_distance = float('inf')
    for i, waypoint in enumerate(path):
        distance = math.hypot(waypoint[0] - robot_position[0], waypoint[1] - robot_position[1])
        if distance < nearest_distance:
            nearest_index = i
            nearest_distance = distance

    # Check if the goal is reached
    if nearest_distance <= goal_radius:
        # Update the goal to the next waypoint or the last waypoint
        goal_index = min(nearest_index + 1, len(path) - 1)
        goal_point = path[goal_index]

        # Check if the last waypoint is reached
        if goal_index == len(path) - 1:
            # Stop motors only if the final goal hasn't been reached yet
            if not hasattr(pure_pursuit, 'final_goal_reached') or not pure_pursuit.final_goal_reached:
                sim.simxSetJointTargetVelocity(clientID, left_motor_handle, 0, sim.simx_opmode_oneshot)
                sim.simxSetJointTargetVelocity(clientID, right_motor_handle, 0, sim.simx_opmode_oneshot)
                print("Reached the final goal!")
                pure_pursuit.final_goal_reached = True  # Set the flag to indicate the final goal is reached
                sys.exit()  # Exit the script when the final goal is reached
            return  # Exit the pure_pursuit function

    else:
        goal_point = path[nearest_index]

    # Calculate the look-ahead point
    look_ahead_point = path[min(nearest_index + 1, len(path) - 1)]

    # Calculate the vector from the robot to the look-ahead point
    look_ahead_vector = [look_ahead_point[0] - robot_position[0], look_ahead_point[1] - robot_position[1]]

    # Calculate the vector from the robot to the current point
    robot_to_current_vector = [goal_point[0] - robot_position[0], goal_point[1] - robot_position[1]]

    # Calculate the angle between the vectors using the dot product
    dot_product = look_ahead_vector[0] * robot_to_current_vector[0] + look_ahead_vector[1] * robot_to_current_vector[1]
    cross_product = look_ahead_vector[0] * robot_to_current_vector[1] - look_ahead_vector[1] * robot_to_current_vector[0]
    alpha = math.atan2(cross_product, dot_product)

    # Set motor velocities
    angular_velocity = alpha * linear_velocity / lookahead_distance
    left_velocity = linear_velocity - angular_velocity * 0.5 * 0.3  # Adjust wheel separation if needed
    right_velocity = linear_velocity + angular_velocity * 0.5 * 0.3
    sim.simxSetJointTargetVelocity(clientID, left_motor_handle, left_velocity, sim.simx_opmode_oneshot)
    sim.simxSetJointTargetVelocity(clientID, right_motor_handle, right_velocity, sim.simx_opmode_oneshot)

    
    # Print debug information
    print(f"robot_yaw: {robot_yaw}, alpha: {alpha}")


# Main loop
while True:
    pure_pursuit()
    time.sleep(0.05)  # Adjust the time interval as needed
```

```python
import sim
import math
import numpy as np

class State():
    def __init__(self, x_, y_, theta_):
        if not (x_ is None or y_ is None or theta_ is None):
            self.x = x_
            self.y = y_
            self.theta = theta_
        else:
            self.x = 0
            self.y = 0
            self.theta = 0

    def __str__(self):
        return str(self.x)+","+str(self.y)+","+str(self.theta)


class Controller():
    def __init__(
            self, start_, goal_, left_motor_handle, right_motor_handle, R_=0.0325, L_=0.1,
            kP=1.0, kI=0.01, kD=0.01, dT=0.1, v=1.0,
            arrive_distance=1):

        self.current = start_
        self.goal = goal_
        self.R = R_  # in meter
        self.L = L_  # in meter

        self.E = 0   # Cummulative error
        self.old_e = 0  # Previous error

        self.Kp = kP
        self.Ki = kI
        self.Kd = kD

        self.desiredV = v
        self.dt = dT  # in second
        self.arrive_distance = arrive_distance

        # Motor handles
        self.left_motor_handle = left_motor_handle
        self.right_motor_handle = right_motor_handle

        return

    def uniToDiff(self, v, w):
        vR = (2 * v + w * self.L) / (2 * self.R)
        vL = (2 * v - w * self.L) / (2 * self.R)
        return vR, vL

    def diffToUni(self, vR, vL):
        v = self.R / 2 * (vR + vL)
        w = self.R / self.L * (vR - vL)
        return v, w

    def move_robot(self, speed, steering_angle):
        wheel_radius = 0.0975  # Replace with your actual wheel radius
        wheel_base = 0.331  # Replace with your actual wheelbase

        left_wheel_speed = (2 * speed - steering_angle * wheel_base) / (2 * wheel_radius)
        right_wheel_speed = (2 * speed + steering_angle * wheel_base) / (2 * wheel_radius)

        # Set the motor speeds
        sim.simxSetJointTargetVelocity(clientID, self.left_motor_handle, left_wheel_speed, sim.simx_opmode_oneshot)
        sim.simxSetJointTargetVelocity(clientID, self.right_motor_handle, right_wheel_speed, sim.simx_opmode_oneshot)

    def iteratePID(self):
        # Difference in x and y
        d_x = self.goal.x - self.current.x
        d_y = self.goal.y - self.current.y

        # Angle from robot to goal
        g_theta = np.arctan2(d_y, d_x)

        # Error between the goal angle and robot angle
        alpha = g_theta - self.current.theta
        e = np.arctan2(np.sin(alpha), np.cos(alpha))

        e_P = e
        e_I = self.E + e
        e_D = e - self.old_e

        # This PID controller only calculates the angular
        # velocity with constant speed of v
        # The value of v can be specified by giving in parameter or
        # using the pre-defined value defined above.
        w = self.Kp * e_P + self.Ki * e_I + self.Kd * e_D

        w = np.arctan2(np.sin(w), np.cos(w))

        self.E = self.E + e
        self.old_e = e
        v = self.desiredV

        return v, w

    def fixAngle(self, angle):
        return np.arctan2(np.sin(angle), np.cos(angle))

    def makeAction(self, v, w):
        x_dt = v * np.cos(self.current.theta)
        y_dt = v * np.sin(self.current.theta)
        theta_dt = w

        self.current.x = self.current.x + x_dt * self.dt
        self.current.y = self.current.y + y_dt * self.dt
        self.current.theta = self.fixAngle(
            self.current.theta + self.fixAngle(theta_dt * self.dt))

        # Move the robot
        self.move_robot(v, w)

    def isArrived(self):
        current_state = np.array([self.current.x, self.current.y])
        goal_state = np.array([self.goal.x, self.goal.y])
        difference = current_state - goal_state

        distance_err = difference @ difference.T
        if distance_err < self.arrive_distance:
            return True
        else:
            return False

    def runPID(self):
        while not self.isArrived():
            v, w = self.iteratePID()
            self.makeAction(v, w)
            # time.sleep(self.dt)

# Connect to CoppeliaSim
sim.simxFinish(-1)  # Close connections from previous sessions
clientID = sim.simxStart('127.0.0.1', 19997, True, True, 5000, 5)

if clientID != -1:
    print('Connected to CoppeliaSim')

    # Get robot handles
    error_code, left_motor_handle = sim.simxGetObjectHandle(clientID, 'Pioneer_p3dx_leftMotor', sim.simx_opmode_oneshot_wait)
    error_code, right_motor_handle = sim.simxGetObjectHandle(clientID, 'Pioneer_p3dx_rightMotor', sim.simx_opmode_oneshot_wait)

    # Define path coordinates
    path = [
        State(-1.75, 2.25, 0),
        State(-1.6, 2.25, 0),
        # ... (add the rest of your path coordinates)
        State(1.75, 0.25, 0)
    ]

    goal_point = path[-1]  # The last point in the path

    # Initial robot pose is the first point in the path with heading 0
    robot_pose = State(path[0].x, path[0].y, 0)

    # Create Controller
    controller = Controller(robot_pose, goal_point, left_motor_handle, right_motor_handle)

    # Run PID
    controller.runPID()

    # Stop the simulation after completing the path
    sim.simxStopSimulation(clientID, sim.simx_opmode_blocking)

    # Close the connection
    sim.simxFinish(clientID)
    print('Connection closed')
else:
    print('Failed to connect to CoppeliaSim')
```
